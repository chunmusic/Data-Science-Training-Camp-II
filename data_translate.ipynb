```python
import pandas as pd
import googletrans
from googletrans import Translator
```

```python
translator = Translator()
translations_cat0 = {}
translations_cat1 = {}

element_cat0 = data_df["category_id0"].unique()
for element in element_cat0:
    translations_cat0[element] = translator.translate(element).text
    
element_cat1 = data_df["category_id1"].unique()
for element in element_cat1:
    translations_cat1[element] = translator.translate(element).text
```

```python
data_df["category_id0"].replace(translations_cat0, inplace = True)
data_df["category_id1"].replace(translations_cat1, inplace = True)
```
